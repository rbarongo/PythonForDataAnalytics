```python
#Add labels
x = np.linspace(-np.pi, np.pi, 1000,endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Sin and Cosine Value')
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#Change colors and line style
x = np.linspace(-np.pi, np.pi, 1000,endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Sin and Cosine Value')
plt.plot(x,y1,color='red', linestyle='--',linewidth=4)
plt.plot(x,y2,color='blue',linestyle='-.',linewidth=.5)
plt.show()
```

```python
#Another sytle changing color and style
x = np.linspace(-np.pi, np.pi, 1000,endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Sin and Cosine Value')
plt.plot(x,y1,'b-.')
plt.plot(x,y2,'r--')
plt.show()
```

```python
#Change font size of labels
x = np.linspace(-np.pi, np.pi, 1000,endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlabel('x-values',fontsize='11', color='black')
plt.ylabel('y-values',fontsize='11', color='black')
plt.title('Sin and Cosine Value',fontsize='15', color='blue')
plt.plot(x,y1,'b-.')
plt.plot(x,y2,'r--')
plt.show()
```

```python
#Add markers and grid
x = np.linspace(-np.pi, np.pi, 50,endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlabel('x-values',fontsize='11', color='black')
plt.ylabel('y-values',fontsize='11', color='black')
plt.title('Sin and Cosine Value',fontsize='15', color='blue')

plt.plot(x,y1,marker='s')
plt.grid()
plt.plot(x,y2,'r--')
plt.show()
```

```python
#Add legend
x = np.linspace(-np.pi, np.pi, 50,endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)
plt.xlabel('x-values',fontsize='11', color='black')
plt.ylabel('y-values',fontsize='11', color='black')
plt.title('Sin and Cosine Value',fontsize='15', color='blue')
plt.plot(x,y1,label='y1')
#plt.grid()
plt.plot(x,y2,label='y2')
plt.legend()
#plt.legend('upper right')
#plt.legend('upper left')
plt.show()
```

#### Subplots

```python
#Subplot of 2 rows and 1 column
#Subplot(2 rows, 1 column)
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
```

```python
cosx = np.cos(x)
sinx = np.sin(x)
#sinx
```

```python
plt.subplot(2,1,1)
plt.plot(x,cosx, color='r')
plt.title('CosX')
plt.grid()

plt.subplot(2,1,2)
plt.plot(x,sinx,color='b')
plt.title('SinX')
plt.grid()
```

```python
plt.subplot(3,2,1)
plt.plot(x,cosx, color='r')
plt.title('CosX')
plt.grid()

plt.subplot(3,2,2)
plt.plot(x,x**2,color='b')
plt.title('X square')
#plt.grid()

plt.subplot(3,2,3)
plt.plot(x,cosx, color='r')
plt.title('CosX')
plt.grid()

plt.subplot(3,2,4)
plt.plot(x,sinx,color='b')
plt.title('SinX')
plt.grid()

plt.subplot(3,2,5)
plt.plot(x,cosx, color='r')
plt.title('CosX')
plt.grid()

plt.subplot(3,2,6)
plt.plot(x,sinx,color='b')
plt.title('SinX')
plt.ylim(-0.5,0.5)
plt.grid()
```

```python
#setting limits or ZOOM IN
#plt.subplot(3,2,6)
plt.plot(x,sinx,color='b')
plt.title('SinX')
plt.ylim(-0.8,0.8)
plt.xlim(-1,0.1)
plt.grid()
```
